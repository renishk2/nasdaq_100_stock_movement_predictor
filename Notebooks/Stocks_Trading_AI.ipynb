```python
# Loading the dataset 

stocks_df = pd.read_csv\
('/Users/renishkanjiyani/Documents/BrainStation/Final Project/nasdaq_stocks_100.csv', sep='\t')
```

```python
stocks_df.head()
```

```python
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'])
stocks_df['Date']
```

```python
stocks_df.info()
```

```python
dummy_variables = pd.get_dummies(stocks_df['Name'])
```

```python
dummy_variables = dummy_variables.astype(int)
```

```python
stocks_df = pd.concat([stocks_df, dummy_variables], axis=1)
```

```python
stocks_df
```

```python
stocks_df = stocks_df.drop(columns='Name', axis=1)
```

```python
stocks_df
```

```python
stocks_df['Result'] = np.where(stocks_df['Close'] - stocks_df['Open'] > 0, 1, 0)
```

```python
stocks_df['Result'].value_counts()
```

```python
stocks_df.info()
```

```python
X = stocks_df.drop(columns=['Result', 'Date'], axis=1)
y = stocks_df['Result']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size=0.2, random_state=42)
```

```python
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()

logistic_regression.fit(X_train, y_train)
```

```python
accuracy = logistic_regression.score(X_test, y_test)
print("Accuracy:", accuracy)
```
